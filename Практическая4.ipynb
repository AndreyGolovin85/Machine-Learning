Для выполнения практической работы будет использоваться датасет, находящийся по ссылке: https://archive.ics.uci.edu/ml/datasets/spambase

Данный датасет состоит из 57 атрибутов:
Последний столбец "spambase.data" указывает, считалось ли электронное письмо спамом (1) или нет (0), то есть незапрошенной коммерческой электронной почтой. Большинство атрибутов указывают на то, часто ли встречалось в письме то или иное слово или символ. Атрибуты run-length (55-57) измеряют длину последовательностей последовательных заглавных букв.
Вот определения атрибутов:

48 непрерывных вещественных [0,100] атрибутов типа word_freq_WORD
= процент слов в электронном письме, которые соответствуют СЛОВУ, то есть 100 * (количество раз, когда СЛОВО появляется в электронном письме) / общее количество слов в электронном письме. "Слово" в данном случае-это любая строка буквенно-цифровых символов, ограниченная не буквенно-цифровыми символами или концом строки.

6 непрерывные реальные [0,100] атрибуты типа char_freq_CHAR]
= процент символов в электронном письме, соответствующих CHAR, т. е. 100 * (количество встреч СИМВОЛОВ) / общее количество символов в электронном письме

1 непрерывный реальный [1,...] атрибут типа capital_run_length_average
= средняя длина непрерывных последовательностей заглавных букв

1 непрерывное целое [1,...] атрибут типа capital_run_length_longest
= длина самой длинной непрерывной последовательности заглавных букв

1 непрерывное целое [1,...] атрибут типа capital_run_length_total
= сумма длины непрерывных последовательностей прописных букв
= общее количество заглавных букв в сообщении электронной почты

1 номинальное {0,1} Class атрибут типа спам
= обозначает, является ли электронная почта считалась спам (1) или нет (0), т. е. нежелательной коммерческой электронной почты.

Количество записей в датасете – 4600 шт. 
Пропущенные значения – Присутствуют.
Категориальные атрибуты – отсутствуют.

Как видим в нашей выборке отсутствуют пропуски.

```python
# Переименуем столбец '1' в 'spam' для удобства дальнейшего использования.
data = data.rename(columns={'1': 'spam'})
data
```

```python
X = data.drop(columns=['spam'])
y = data.spam
 
# Разделим набор данных на обучающую и тестовую выборку в соотношении 70% на 30%.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Выберем модель для обучения GradientBoostingClassifier и запустим
процесс обучения.

```python
from sklearn.ensemble import GradientBoostingClassifier
gradient_boosting = GradientBoostingClassifier(learning_rate=0.1,
                                               n_estimators=100, max_depth=3, 
                                               random_state=None, ccp_alpha=0.0)
gradient_boosting.fit(X_train, y_train)
```

Выведем параметры точности обученной модели:

```python
%%time
# Выведем основные параметры точности обученной модели.
from sklearn.metrics import confusion_matrix, classification_report
y_pred_gradient = gradient_boosting.predict(X_test)
confmat_gradient = confusion_matrix(y_true=y_test, y_pred=y_pred_gradient)

print(confmat_gradient)
print("Classification Report GradientBoosting")
print(classification_report(y_test, y_pred_gradient))
```

Выберем модель для обучения RigeClassifier и запустим процесс обучения.

```python
from sklearn.linear_model import RidgeClassifier
clf = RidgeClassifier(alpha=1.0, fit_intercept=True, normalize=False,
                      copy_X=True, max_iter=None, tol=0.001,
                      class_weight=None, solver='auto', random_state=None)
clf.fit(X_train, y_train)
```

Выведем параметры точности обученной модели:

```python
%%time
y_pred_clf = clf.predict(X_test)
confmat_clf = confusion_matrix(y_true=y_test,
y_pred=y_pred_clf)
print(confmat_clf)
print("Classification Report RidgeClassifier")
print(classification_report(y_test, y_pred_clf))
```

Выберем модель для обучения PassiveAggressiveClassifier и запустим процесс обучения.

```python
from sklearn.linear_model import PassiveAggressiveClassifier
clf_pa = PassiveAggressiveClassifier()
clf_pa.fit(X_train, y_train)
```

Выведем параметры точности обученной модели:

```python
%%time
y_pred_clf_pa = clf_pa.predict(X_test)
confmat_clf_pa = confusion_matrix(y_true=y_test,
y_pred=y_pred_clf_pa)
print(confmat_clf_pa)
print("Classification Report PassiveAggressive")
print(classification_report(y_test, y_pred_clf_pa))
```

Вывод. Как видно из отчетов Classification Report GradientBoosting, Classification Report RidgeClassifier, Classification Report PassiveAggressive, наилучшие точностные показатели обеспечивает классификатор, построенный на основе GradientBoostingClassifier.
