```python
import numpy as np
import pandas as pd

#Создаем датафрейм. Столбцы a1...a3, b1...b3, c1...c3 содержат значения компонентов 
#квадратной матрицы раззмером 3x3. Столбец det содержит значение детерминанта матрицы.

df_input = pd.DataFrame(columns=['a1', 'a2', 'a3', 'b1', 'b2', 'b3', 'c1', 'c2',
                                 'c3', 'det'])
df_input
```

```python
# Внутри цикла генерируем обучающую выборку с помощью функции randint().
for i in range(2000):
    a1 = np.random.randint(0, 11)
    a2 = np.random.randint(0, 11)
    a3 = np.random.randint(0, 11)
    b1 = np.random.randint(0, 11)
    b2 = np.random.randint(0, 11)
    b3 = np.random.randint(0, 11)
    c1 = np.random.randint(0, 11)
    c2 = np.random.randint(0, 11)
    c3 = np.random.randint(0, 11)

# Расчет детерминанта det случайно сгенерированной матрицы производим с помощью
# функции linalg.det()
    det = np.linalg.det(np.array([[a1, a2, a3],[b1, b2, b3],[c1, c2, c3]]))
    df_input.loc[i] = [a1, a2, a3, b1, b2, b3, c1, c2, c3, det]
df_input
```

```python
# Формируем массив X и вектор y.
X = df_input.drop(columns=['det'])  # В мвссив X помещаем все столбцы кроме det.
y = df_input.det                    # В вектор y помещаем только det.
X
```

```python
#@title Текст заголовка по умолчанию
pd.DataFrame(y)
```

```python
#Подключаем функцию для случайного разделения исходных данных
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01,
                                                    random_state=9)

X_train
```

```python
%%time
#В качастве нейронной сети выбираем многослойный персептрон по умолчанию он
#содержит один скрытый слой со 100 нейронами.
from sklearn.neural_network import MLPRegressor
model=MLPRegressor()

#Запускаем процесс оббучения и подсчитывваем время затраченное для настройки
#нейронной сети.
model.fit(X_train, y_train)
```

```python
#Весовые коэффициенты всех слоев
model.coefs_
```

```python
#Определение точности работы нейронной сети

R2 = model.score(X_test, y_test)
print('R2='+str(R2))
```
